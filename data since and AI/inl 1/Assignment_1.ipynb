```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

data = pd.read_csv("life-expectancy-vs-gdp-per-capita.csv")


data = data.dropna(subset = ["Life expectancy", "GDP per capita"])
data = data[(data["Year"] == 2018) & (data["Code"].str.count("") <=4) ]



y = data["Life expectancy"]

x = data["GDP per capita"]

x.describe()
data
```

```python
plt.scatter(y,x,marker='.',s=2)
plt.title("Life expectancy vs GDP per capita")
plt.xlabel("Life expectancy	")
plt.ylabel("GDP per capita")
```

```python
# Happiness

df4 = pd.read_csv("gdp-vs-happiness.csv")
df4 = df4.dropna(subset = ["Life satisfaction in Cantril Ladder (World Happiness Report 2022)", "GDP per capita, PPP (constant 2017 international $)"])
df4 = df4[(df4["Year"] == 2018)]

plt.scatter(df4["GDP per capita, PPP (constant 2017 international $)"],df4["Life satisfaction in Cantril Ladder (World Happiness Report 2022)"],marker='.', s=2)
plt.title("Happiness vs GDP per capita")
plt.xlabel("GDP per capita")
plt.ylabel("Happiness")
```

```python
# Alcohol 
Alcohol = pd.read_csv("alcohol-consumption-vs-gdp-per-capita.csv")
Alcohol = Alcohol.dropna(subset = ["Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)", "GDP per capita, PPP (constant 2017 international $)"])
Alcohol = Alcohol[(Alcohol["Year"] == 2018)]
plt.scatter(Alcohol["GDP per capita, PPP (constant 2017 international $)"],Alcohol["Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)"],marker='.', s=2)
plt.title("alcohol consumption vs gdp per capita")
plt.xlabel("GDP per capita")
plt.ylabel("alcohol consumption ")



```

```python
#  Population density per gdp

pop_densiti = pd.read_csv("population-density-vs-prosperity.csv")
pop_densiti = pop_densiti.dropna(subset = ["population_density", "GDP per capita, PPP (constant 2017 international $)"])

pop_densiti = pop_densiti[(pop_densiti["Year"] == 2018)]

plt.scatter(pop_densiti["GDP per capita, PPP (constant 2017 international $)"],pop_densiti["population_density"],marker='.', s=2)
plt.title("Happiness vs GDP per capita")
plt.xlabel("GDP per capita")
plt.ylabel("Happiness")
```

```python
democ = pd.read_csv("gdp-per-capita-vs-liberal-democracy.csv")
democ = democ.dropna(subset = ["libdem_vdem_owid", "GDP per capita"])

democ = democ[(democ["Year"] == 2018)]

plt.scatter(democ["GDP per capita"],democ["libdem_vdem_owid"],marker='.', s=2)
plt.title("gdp per capita vs liberal democracy")
plt.xlabel("GDP per capita")
plt.ylabel("liberal democracy")
```
